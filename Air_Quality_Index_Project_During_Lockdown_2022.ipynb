Extracting air quality information

**Plotting pollutants graph**

```python
import matplotlib.pyplot as plt

pollutants = [i for i in iaqi]
values = [i['v'] for i in iaqi.values()]


# Exploding the first slice
explode = [0 for i in pollutants]
mx = values.index(max(values))  # explode 1st slice
explode[mx] = 0.1

# Plot a pie chart
plt.figure(figsize=(8,6))
plt.pie(values, labels=pollutants,explode=explode,autopct='%1.1f%%', shadow=True)

plt.title('Air pollutants and their probable amount in atmosphere [Delhi]')

plt.axis('equal')
plt.show()
```

Read csv files into a dataframe

```python
import pandas as pd


csv_path = '/content/drive/MyDrive/Colab Notebooks DataSets/Air Quality Data Analysis Project/r.k.-puram, delhi, delhi, india-air-quality.csv'

df = pd.read_csv(csv_path)
df = df.head(50)
print(df)
```

```python
df
```

```python
df.columns

df= df.rename(columns = {" pm25": "pm25", 
                         " pm10":"pm10", 
                         " o3": "o3",
                         ' no2' : 'no2',
                         ' so2' : 'so2',
                         ' co' : 'co'})

df.columns
```

### Extracting 21 days after lockdown

```python
df['date'] = pd.to_datetime(df.date)

df22 = df.loc[df['date'] > '2022-02-04']
df22 = df22.sort_values(by = 'date')
df22
```

```python
len(df22)
```

**Remove any extra date**

```python
df22.drop(25, inplace=True)
df22
```

```python
len(df22)
```

**Filling all the empty cells with 0**

```python
df22.replace(' ', '0', inplace=True)
df22
```

**Plotting the data**

```python
import matplotlib.pyplot as plt

dates = df22['date']
pm25 = df22['pm25']
pm25 = [int(i) for i in pm25]

plt.figure(figsize=(10,8))

length = len(dates)

plt.plot(dates,pm25, color = 'green')
plt.title('PM2.5 values after lockdown days')
plt.xlabel('Dates after 21 days lockdown')
plt.ylabel('PM2.5 values')
plt.show()
```

**Extracting 21 days under lockdown**

```python
mask = (df['date'] >= '2022-01-08') & (df['date']  < '2022-02-03')

past22 = df.loc[mask]
past22
```

```python
len(past22)
```

### **Compairing above two**

```python
import matplotlib.pyplot as plt

dates = df22['date']
pm25_l = df22['pm25']
pm25_l = [int(i) for i in pm25]


pm25_n = past22['pm25']
pm25_n = [int(i) for i in pm25_n]

plt.figure(figsize=(10,8))

length = [i for i in range(1,len(dates)+1)]

plt.plot(length,pm25_l,color='green',label='After 21 days lockdown')
plt.plot(length,pm25_n,color='red',label='Under lockdown')
plt.legend()
plt.title('Comparision of before lockdown vs under lockdown pm2.5 values')
plt.show()
```

**Finding solutions to several questions**

1. On which date pm2.5 value was minimum in Delhi under lockdown

```python
df22['pm25'] = [int(i) for i in df22['pm25']]
print(df22[df22.pm25 == df22.pm25.min()]) 
```

2. On which date o3 value was maximum in Delhi under lockdown

```python
df22['o3'] = [int(i) for i in df22['o3']]
print(df22[df22.o3 == df22.o3.max()]) 
```

3. What is the average value of so2 in the lockdown

```python
df22['so2'] = [int(i) for i in df22['so2']]
avgSo2 = df22['so2'].mean()
print('The average value of so2 :',avgSo2)
```
